```python
import pandas as pd
import geocoder
```

```python
file = pd.read_excel("/Users/verafeng/Desktop/LCT_LancasterCountyTax/LancasterCountyTaxAccountList_coordinates.xlsx")
```

```python
start = 100000
gap = 10000
file.shape
```

```python
address = []
print ('Start from ' + str(start+1) + ' to ' + str(start+gap))
for x in range(start, start+gap):
    address.append(file["address"][x])
address
len(address)
```

```python
cnt = start
coordinates = []
for i in address:
    g = geocoder.arcgis(i)
    cnt = cnt + 1
#     print (i + '\n')
    print(str(cnt)+ ':' + str(g.latlng) + '\n')
    coordinates.append(str(g.latlng))
```

```python
coordinates
```

```python
df = file.iloc[start:start+gap,:]

df['coordinates'] = coordinates
```
